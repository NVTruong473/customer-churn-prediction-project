## **DATA CLEANING**

```python
# DATA CLEANING

# Create copy
df_clean = df.copy()

# Drop customerID
df_clean.drop("customerID", axis=1, inplace=True)

# Replace blank values in TotalCharges
df_clean["TotalCharges"] = df_clean["TotalCharges"].replace(" ", np.nan)

# Convert to numeric
df_clean["TotalCharges"] = pd.to_numeric(df_clean["TotalCharges"])

# Fill missing values with median
df_clean["TotalCharges"].fillna(df_clean["TotalCharges"].median(), inplace=True)

# Convert target column
df_clean["Churn"] = df_clean["Churn"].map({"Yes": 1, "No": 0})

print("Data cleaning completed!")
print("New Shape:", df_clean.shape)

display(df_clean.head())
```

### **INSIGHT OF DATA CLEANING**

- Customer identifiers were removed because they do not provide predictive value.   (Mã khách hàng đã được xóa vì không mang giá trị dự đoán.)

- TotalCharges was converted into numeric format for proper analysis.   (TotalCharges đã được chuyển sang dạng số để phân tích đúng cách.)

- Missing values were filled using the median to reduce distortion.  (Giá trị thiếu được điền bằng median để giảm sai lệch.)

- The target column was encoded into binary labels for Machine Learning models.  (Cột mục tiêu đã được mã hóa nhị phân để dùng cho mô hình Machine Learning.)

## **PREPROCESSING**

### **Preprocessing Insight (Nhận Xét Tiền Xử Lý)**

- Categorical variables were successfully converted into numeric features.  
(Biến phân loại đã được chuyển thành đặc trưng số.)

- Feature scaling was applied to improve model performance.  
(Chuẩn hóa dữ liệu đã được áp dụng để cải thiện hiệu suất mô hình.)

```python
# PREPROCESSING

# Features and target
X = df_clean.drop("Churn", axis=1)
y = df_clean["Churn"]

# One-hot encoding
X = pd.get_dummies(X, drop_first=True)

# Save column names
feature_names = X.columns

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Preprocessing completed!")
print("Feature Shape:", X_scaled.shape)
print("Target Shape:", y.shape)
```

## **TRAIN / TEST SPLIT**

```python
# TRAIN TEST SPLIT

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("Train/Test split completed!")
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# VISUAL CHECK OF TRAIN / TEST

fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Train target distribution
sns.countplot(x=y_train, ax=axes[0])
axes[0].set_title("Train Set Target Distribution")
axes[0].set_xlabel("Churn")

# Test target distribution
sns.countplot(x=y_test, ax=axes[1])
axes[1].set_title("Test Set Target Distribution")
axes[1].set_xlabel("Churn")

plt.tight_layout()
plt.show()
```

## **MODEL TRAINING**

### **A PRIMITIVE BEGINNING**

```python
from sklearn.svm import SVC
from sklearn.metrics import precision_score, recall_score, f1_score
```

```python
# ADVANCED MODEL COMPARISON

models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(probability=True)
}

results = []

for name, model in models.items():
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    pre = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    results.append([
        name,
        round(acc, 4),
        round(pre, 4),
        round(rec, 4),
        round(f1, 4),
        round(auc, 4)
    ])

results_df = pd.DataFrame(
    results,
    columns=["Model", "Accuracy", "Precision", "Recall", "F1-score", "ROC-AUC"]
)

results_df = results_df.sort_values(by="ROC-AUC", ascending=False)

display(results_df)
```

```python
# MODEL RANKING CHART

plt.figure(figsize=(10,5))

sns.barplot(data=results_df, x="ROC-AUC", y="Model")
plt.title("Model Comparison by ROC-AUC")

plt.xlim(0.75, 1.00)
plt.show()
```

### **HYPERPARAMETER TUNING**

```python
# ADVANCED HYPERPARAMETER TUNING

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# Random Forest Tuning
rf_params = {
    "n_estimators": randint(100, 500),
    "max_depth": [5, 10, 15, 20, None],
    "min_samples_split": randint(2, 10),
    "min_samples_leaf": randint(1, 5),
    "max_features": ["sqrt", "log2", None]
}

rf_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=rf_params,
    n_iter=25,
    scoring="roc_auc",
    cv=5,
    random_state=42,
    n_jobs=-1
)

rf_search.fit(X_train, y_train)

print("Best Random Forest Params:")
print(rf_search.best_params_)
print("Best RF CV ROC-AUC:", round(rf_search.best_score_, 4))


# SVM Tuning
svm_params = {
    "C": uniform(0.1, 10),
    "kernel": ["linear", "rbf"],
    "gamma": ["scale", "auto", 0.01, 0.001]
}

svm_search = RandomizedSearchCV(
    estimator=SVC(probability=True),
    param_distributions=svm_params,
    n_iter=20,
    scoring="roc_auc",
    cv=5,
    random_state=42,
    n_jobs=-1
)

svm_search.fit(X_train, y_train)

print("\nBest SVM Params:")
print(svm_search.best_params_)
print("Best SVM CV ROC-AUC:", round(svm_search.best_score_, 4))
```

### **RECOMPARE TUNED MODELS VS LOGISTIC REGRESSION**

```python
# FINAL MODEL COMPARISON

# Baseline Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)

# Tuned Models
rf_model = rf_search.best_estimator_
svm_model = svm_search.best_estimator_

models = {
    "Logistic Regression": log_model,
    "Tuned Random Forest": rf_model,
    "Tuned SVM": svm_model
}

results = []

for name, model in models.items():

    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    acc = accuracy_score(y_test, y_pred)
    pre = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_prob)

    results.append([
        name,
        round(acc, 4),
        round(pre, 4),
        round(rec, 4),
        round(f1, 4),
        round(auc, 4)
    ])

final_df = pd.DataFrame(
    results,
    columns=["Model", "Accuracy", "Precision", "Recall", "F1-score", "ROC-AUC"]
)

final_df = final_df.sort_values(by="ROC-AUC", ascending=False)

display(final_df)
```

```python
# FINAL RANKING CHART

plt.figure(figsize=(10,5))

sns.barplot(data=final_df, x="ROC-AUC", y="Model")
plt.title("Final Model Ranking (ROC-AUC)")

plt.xlim(0.75, 1.00)
plt.show()
```

## **PREDICT NEW CUSTOMER (DỰ ĐOÁN KHÁCH HÀNG MỚI)**

```python
# CONNECTED CUSTOMER CHURN REPORT

from IPython.display import display, HTML

# Select 10 random customers
sample_idx = np.random.choice(df_clean.index, 10, replace=False)

sample_info = df_clean.loc[sample_idx].copy()
sample_X = X.loc[sample_idx]
sample_scaled = scaler.transform(sample_X)

# Predict
pred = rf_search.best_estimator_.predict(sample_scaled)
prob = rf_search.best_estimator_.predict_proba(sample_scaled)[:, 1]

# Risk rules
def risk_level(p):
    if p < 0.30:
        return "Low Risk"
    elif p < 0.60:
        return "Medium Risk"
    else:
        return "High Risk"

# Build result table
result_df = pd.DataFrame({
    "Tenure": sample_info["tenure"],
    "Contract": sample_info["Contract"],
    "Monthly Charges": sample_info["MonthlyCharges"],
    "Payment Method": sample_info["PaymentMethod"],
    "Prediction": ["Churn" if x == 1 else "Stay" for x in pred],
    "Probability": np.round(prob, 4),
    "Risk Level": [risk_level(x) for x in prob],
    "Actual": ["Churn" if x == 1 else "Stay" for x in y.loc[sample_idx]]
})

result_df = result_df.sort_values(by="Probability", ascending=False)

# Styling function
def color_risk(val):
    if val == "High Risk":
        return "background-color:#f8d7da; color:#721c24;"
    elif val == "Medium Risk":
        return "background-color:#fff3cd; color:#856404;"
    else:
        return "background-color:#d4edda; color:#155724;"

def color_pred(val):
    if val == "Churn":
        return "color:#b30000; font-weight:bold;"
    return "color:#006400; font-weight:bold;"

styled = (
    result_df.style
    .applymap(color_risk, subset=["Risk Level"])
    .applymap(color_pred, subset=["Prediction", "Actual"])
    .format({"Probability": "{:.2%}"})
    .set_properties(**{
        "text-align": "center",
        "border": "1px solid #dddddd"
    })
    .set_table_styles([
        {"selector": "th",
         "props": [("background-color", "#1f4e78"),
                   ("color", "white"),
                   ("text-align", "center"),
                   ("padding", "8px")]},
        {"selector": "td",
         "props": [("padding", "6px")]}
    ])
)

# Summary counts
high_count = (result_df["Risk Level"] == "High Risk").sum()
med_count = (result_df["Risk Level"] == "Medium Risk").sum()
low_count = (result_df["Risk Level"] == "Low Risk").sum()

# Report Header
display(HTML("""
<h2 style='color:#1f4e78;'>Customer Churn Prediction Report</h2>
<p>This report predicts churn risk for 10 customers using the tuned Random Forest model.</p>

<h3 style='color:#1f4e78;'>How to Read This Report</h3>
<ul>
<li><b>Probability</b>: Chance that customer will leave.</li>
<li><b>Low Risk</b>: 0% - 29%</li>
<li><b>Medium Risk</b>: 30% - 59%</li>
<li><b>High Risk</b>: 60% - 100%</li>
<li><b>Prediction</b>: Model output.</li>
<li><b>Actual</b>: Real label from dataset.</li>
</ul>

<h3 style='color:#1f4e78;'>Risk Summary</h3>
<ul>
<li>High Risk Customers: """ + str(high_count) + """</li>
<li>Medium Risk Customers: """ + str(med_count) + """</li>
<li>Low Risk Customers: """ + str(low_count) + """</li>
</ul>

<h3 style='color:#1f4e78;'>Detailed Results</h3>
"""))

display(styled)
```
